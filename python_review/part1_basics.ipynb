## FizzBuzz

Create a while loop that goes through integers of 1 to 50 and prints "Fizz" for multiples of 3, and "Buzz" for multiples of "5", and "FizzBuzz" for multiples of 3 and 5.

```python
# Code here
i = 1
while i <= 50:
    if i % 3 == 0 and i % 5 == 0:
        print("Fizzbuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
    i += 1
```
